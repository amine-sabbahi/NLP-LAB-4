## 1 - Collecting Arabic text data from several Arabic web site using BeautifulSoup :

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def length_based_score(text):
    # Normalize the score based on the length of the text
    score = (len(text) + 5793) % 10
    return score

def scrape_and_prepare_dataset(urls, score_function=length_based_score):
    dataset = []

    for url in urls:
        try:
            response = requests.get(url)
            response.raise_for_status()  # Check if the request was successful
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract text data (modify according to the website's structure)
            texts = [p.text for p in soup.find_all('p')]
            if texts:
                score = score_function(texts)
                dataset.append({"text": texts, "score": score})


        except requests.exceptions.RequestException as e:
            print(f"Error scraping {url}: {e}")

    # Convert dataset to DataFrame for easy manipulation
    df = pd.DataFrame(dataset)
    return df
```

## 2-  Establishing a preprocessing NLP pipeline (Tokenization stemming lemmatization, Stop words) for the collected Dataset:

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.isri import ISRIStemmer

# Text Cleaning
def clean_text(text):
    if isinstance(text, list):
        text = " ".join(text)
    text = re.sub(r'<.*?>', '', text)  # Removing HTML tags
    text = re.sub(r'[^\w\s]', '', text)  # Removing punctuation and numbers
    text = re.sub(r'[a-zA-Z0-9]', '', text)  # Removing English letters and numbers
    text = re.sub(r'[\n\r]+', ' ', text)  # Removing new lines and carriage returns
    text = re.sub(r'\s+', ' ', text)  # Collapsing multiple spaces
    return text.lower()

# Tokenization
def tokenize_text(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
    else:
        tokens = []
    return tokens

# Stop Words Removal
arabic_stop_words = set(stopwords.words('arabic'))  # Define your own list if needed

def remove_stop_words(tokens):
    if isinstance(tokens, list):
        filtered_tokens = [word for word in tokens if word not in arabic_stop_words]
    else:
        filtered_tokens = []
    return filtered_tokens

# Normalization (optional)
def normalize_text(tokens):
    if isinstance(tokens, list):
        normalized_tokens = [word.lower() for word in tokens]
    else:
        normalized_tokens = []
    return normalized_tokens

# Stemming
def stemming(tokens):
    stemmer = ISRIStemmer()
    if isinstance(tokens, list):
        stemmed_tokens = [stemmer.stem(token) for token in tokens]
    else:
        stemmed_tokens = []
    return stemmed_tokens

# Lemmatization
def lemmatization(tokens):
    lemmatizer = WordNetLemmatizer()
    if isinstance(tokens, list):
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    else:
        lemmatized_tokens = []
    return lemmatized_tokens

# Complete preprocessing function
def preprocess_text(text):
    cleaned_text = clean_text(text)
    tokens = tokenize_text(cleaned_text)
    tokens = remove_stop_words(tokens)
    tokens = normalize_text(tokens)
    tokens = stemming(tokens)
    tokens = lemmatization(tokens)
    return tokens
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Assume df is your DataFrame with 'Processed_Text' and 'Score' columns

# Example preprocessed tokens
df['Processed_Text'] = df['text'].apply(preprocess_text)

# Create vocabulary and convert text to indices
vocab = set(token for tokens in df['Processed_Text'] for token in tokens)
word_to_idx = {word: idx for idx, word in enumerate(vocab)}

def text_to_indices(tokens):
    return [word_to_idx[token] for token in tokens if token in word_to_idx]

df['Text_Indices'] = df['Processed_Text'].apply(text_to_indices)

# Example padding function (assuming max length of 10 for simplicity)
def pad_sequence(seq, max_length=10):
    if len(seq) < max_length:
        seq.extend([0] * (max_length - len(seq)))
    return seq[:max_length]

df['Padded_Indices'] = df['Text_Indices'].apply(pad_sequence)

# Update the TextDataset class to handle DataFrame
class TextDataset(Dataset):
    def __init__(self, texts, scores):
        self.texts = texts
        self.scores = scores

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return torch.tensor(self.texts[idx], dtype=torch.long), torch.tensor(self.scores[idx], dtype=torch.float)

# Create dataset and dataloader
texts = df['Padded_Indices'].tolist()
scores = df['score'].tolist()
dataset = TextDataset(texts, scores)
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

# Verify DataLoader output
for batch in dataloader:
    print(batch)
    break
```

```python
import torch.nn as nn

class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(RNNModel, self).__init__()
        self.embedding = nn.Embedding(len(vocab), input_size)
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.embedding(x)
        h0 = torch.zeros(num_layers, x.size(0), hidden_size).to(device)
        out, _ = self.rnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out

class BiRNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(BiRNNModel, self).__init__()
        self.embedding = nn.Embedding(len(vocab), input_size)
        self.birnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, output_size)

    def forward(self, x):
        x = self.embedding(x)
        h0 = torch.zeros(num_layers * 2, x.size(0), hidden_size).to(device)
        out, _ = self.birnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out

class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(GRUModel, self).__init__()
        self.embedding = nn.Embedding(len(vocab), input_size)
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.embedding(x)
        h0 = torch.zeros(num_layers, x.size(0), hidden_size).to(device)
        out, _ = self.gru(x, h0)
        out = self.fc(out[:, -1, :])
        return out

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(LSTMModel, self).__init__()
        self.embedding = nn.Embedding(len(vocab), input_size)
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.embedding(x)
        h0 = torch.zeros(num_layers, x.size(0), hidden_size).to(device)
        c0 = torch.zeros(num_layers, x.size(0), hidden_size).to(device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
# Training loop
def train_model(model, dataloader, num_epochs=10, learning_rate=0.001):
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    model.train()
    for epoch in range(num_epochs):
        for texts, scores in dataloader:
            texts, scores = texts.to(device), scores.to(device)

            outputs = model(texts)
            loss = criterion(outputs, scores)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Initialize and train models
input_size = 50  # Example input size, modify as needed
hidden_size = 50
output_size = 1
num_layers = 1
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

rnn_model = RNNModel(input_size, hidden_size, output_size, num_layers).to(device)
train_model(rnn_model, dataloader)

birnn_model = BiRNNModel(input_size, hidden_size, output_size, num_layers).to(device)
train_model(birnn_model, dataloader)

gru_model = GRUModel(input_size, hidden_size, output_size, num_layers).to(device)
train_model(gru_model, dataloader)

lstm_model = LSTMModel(input_size, hidden_size, output_size, num_layers).to(device)
train_model(lstm_model, dataloader)

# Evaluate models
def evaluate_model(model, dataloader):
    model.eval()
    predictions = []
    actuals = []
    with torch.no_grad():
        for texts, scores in dataloader:
            texts, scores = texts.to(device), scores.to(device)
            outputs = model(texts).cpu().numpy()
            predictions.extend(outputs)
            actuals.extend(scores.cpu().numpy())

    mse = mean_squared_error(actuals, predictions)
    mae = mean_absolute_error(actuals, predictions)
    r2 = r2_score(actuals, predictions)
    print(f"Mean Squared Error: {mse:.4f}")
    print(f"Mean Absolute Error: {mae:.4f}")
    print(f"R-squared: {r2:.4f}")

    return mse, mae, r2

# Example: Evaluate models
print("\n Evaluating RNN Model: \n")
mse_rnn, mae_rnn, r2_rnn = evaluate_model(rnn_model, dataloader)

print("\nEvaluating Bidirectional RNN Model: \n")
mse_birnn, mae_birnn, r2_birnn = evaluate_model(birnn_model, dataloader)

print("\nEvaluating GRU Model: \n")
mse_gru, mae_gru, r2_gru = evaluate_model(gru_model, dataloader)

print("\nEvaluating LSTM Model: \n")
mse_lstm, mae_lstm, r2_lstm = evaluate_model(lstm_model, dataloader)
```

## 1 Fine tuning the pre-trained model (GPT2) to a Midjoureny prompts image generation Dataset:

```python
import pandas as pd
from transformers import GPT2Tokenizer, GPT2LMHeadModel, AdamW, get_linear_schedule_with_warmup
from torch.utils.data import DataLoader, Dataset
import torch


# Load the dataset
Prompts = pd.read_csv('/kaggle/input/midjourney-image-generation-prompts/midjourney_prompts.csv')
inputs = Prompts['Input'].tolist()
outputs = Prompts['Output'].tolist()
Prompts.head(10)
```

```python
# Initialize the tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Set the padding token to be the eos_token
tokenizer.pad_token = tokenizer.eos_token

# Verify that padding token is set
assert tokenizer.pad_token is not None
```

```python
class TextDataset(Dataset):
    def __init__(self, inputs, outputs, tokenizer, max_length):
        self.inputs = inputs
        self.outputs = outputs
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        input_text = self.inputs[idx]
        output_text = self.outputs[idx]
        # Concatenate input and output texts
        combined_text = f"{input_text} {self.tokenizer.eos_token} {output_text}"
        inputs = self.tokenizer.encode_plus(
            combined_text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            return_tensors='pt',
            truncation=True
        )
        input_ids = inputs['input_ids'].squeeze()
        attention_mask = inputs['attention_mask'].squeeze()        
        return input_ids, attention_mask
```

```python
# Parameters
max_length = 50
batch_size = 8  # Adjust based on your GPU memory
epochs = 6
learning_rate = 5e-5

# Create dataset and dataloader
dataset = TextDataset(inputs, outputs, tokenizer, max_length)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
# Initialize the model
model = GPT2LMHeadModel.from_pretrained('gpt2')
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Initialize the optimizer and scheduler
optimizer = AdamW(model.parameters(), lr=learning_rate)
total_steps = len(dataloader) * epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)
```

```python
# Fine-tuning loop
model.train()
for epoch in range(epochs):
    total_loss = 0
    for batch in dataloader:
        input_ids, attention_mask = batch
        input_ids, attention_mask = input_ids.to(device), attention_mask.to(device)

        outputs = model(input_ids, attention_mask=attention_mask, labels=input_ids)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(dataloader)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")
```

```python
model.eval()

def generate_text(prompt, max_length=50):
    inputs = tokenizer.encode(prompt, return_tensors='pt').to(device)
    outputs = model.generate(
        inputs,
        max_length=max_length,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        early_stopping=True
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def generate_multiple_prompts(prompts, max_length=50):
    generated_texts = []
    for prompt in prompts:
        generated_text = generate_text(prompt, max_length)
        generated_texts.append(generated_text)
    return generated_texts
```

### 1. establishing the pre-trained bert-base-uncased model and preparing the data:


```python
import pandas as pd

# Load dataset
dataset_path = "/kaggle/input/amazon-review-dataset/AMAZON_FASHION_5.json"
data = pd.read_json(dataset_path, lines=True)

# Display sample data
data.head()
```

```python
from sklearn.model_selection import train_test_split

# Select relevant columns and clean data
data = data[['reviewText', 'overall']].dropna()
data['reviewText'] = data['reviewText'].astype(str)
# Adjust labels to be in the range [0, 4]
data['overall'] = data['overall'] - 1
# Split the dataset
train_texts, test_texts, train_labels, test_labels = train_test_split(
    data['reviewText'].tolist(),
    data['overall'].tolist(),
    test_size=0.2,
    random_state=42
)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the data
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=512)
test_encodings = tokenizer(test_texts, truncation=True, padding=True, max_length=512)
```

```python
import torch

class AmazonReviewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = AmazonReviewsDataset(train_encodings, train_labels)
test_dataset = AmazonReviewsDataset(test_encodings, test_labels)
```

### 3. Fine tuning and Training your model

```python
from transformers import BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import accuracy_score, f1_score

model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=5)

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)

# Define compute_metrics function
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='weighted')
    return {
        'accuracy': acc,
        'f1': f1,
    }

# Initialize Trainer
trainer = Trainer(
    model=model,                        
    args=training_args,                  
    train_dataset=train_dataset,         
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)

# Train the model
trainer.train()
```

### 4. Evaluating the model by using standard metrics (Accuracy, Loss, F1 score):

```python
# Evaluate the model and print metrics
eval_result = trainer.evaluate()

# Extract and print accuracy and F1 score
accuracy = eval_result['eval_accuracy']
f1 = eval_result['eval_f1']

print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
```
